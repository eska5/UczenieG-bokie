##**Zadanie 2.** Wykorzystanie splotu rnych jder na obrazach



Warstwy splotowe same ucz si wyodrbniania istotnych cech z danych - lokalnych wzorcw. Wzorce te s rozpoznawane przez sie niezalenie od ich pooenia. Wagi poszczeglnych neuronw reprezentowane s przez jdra splotowe (filtry), ktrych rozmiar jest taki jak rozmiar pola recepcyjnego. Wybr wartoci danego jdra determinuje to, jakie wzorce bd wykrywane.

**2a**. Wykorzystujc funkcj *apply_conv2d*, ktra pozwala na obliczenie splotu jdra z obrazem wejciowym, przeprowad eksperymenty wykorzystujc zdefiniowane wczeniej jdra oraz obraz *lab1_1.jpg*. Sprawd i wywietl jakie cechy podkrelaj dla obrazu w skali szaroci poszczeglne filtry - wywietl obraz przed i po zastosowaniu splotu. 
Dodatkowo, stwrz jeden wasny filtr 5x5 i rwnie wywietl efekt splotu dla tego jdra.

```python
import numpy as np
import tensorflow as tf

def apply_conv2d(input, kernel, strides=(1,1), padding='SAME'):
 
  #zmieniamy ksztat danych na (batch, no_of_rows, no_of_columns, no_of_channels)
  input = input.astype(np.float64)
  input = input[np.newaxis, ...]

  #wymiary jdra
  k_width, k_height = kernel.shape
  #liczba wyj
  no_of_filters = 1
  
  input_depth = 1

  #ksztat jdra powinnien by: (kernel_width, kernel_height, input_depth, no_of_filters)
  k_shape=(k_width, k_height, input_depth, no_of_filters)

  kernel_c = kernel.reshape(k_shape)

  #tworzymy warstw splotow 2D
  layer_2D_1 = tf.keras.layers.Conv2D(no_of_filters, 
                                 kernel_size=(k_width, k_height), 
                                 strides = strides,
                                 padding=padding,
                                 activation='linear', 
                                 use_bias=False, 
                                 weights=[kernel_c])
  
  return layer_2D_1(input).numpy()
```

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# normalizacja danych 0-1
x_train = x_train / 255.0
x_test = x_test / 255.0

print("Ksztat danych treningowych: ", x_train.shape)
```

W kolejnym kroku zdefiniujmy podstawowe parametry treningu.

```python
# wybr parametrw uczenia
batch_size = 64
epochs = 15
no_of_classes = 10
learning_rate = 0.01
```

Po przygotowaniu danych, wykorzystujc API Keras, zdefiniujmy prosty model oraz wywietlmy architektur stworzonej sieci. 

```python
# definicja modelu
keras_model = tf.keras.Sequential([
    
    tf.keras.layers.Conv2D(2, kernel_size=4, activation='relu', 
                           input_shape=x_train.shape[1:], padding='SAME'),
    tf.keras.layers.Flatten(),  # spaszczenie danych do wykorzystania warstwy gstej
    tf.keras.layers.Dense(50, activation='relu'),
    tf.keras.layers.Dense(no_of_classes, activation='softmax')
    
  ])

# wywietlenie architektury
keras_model.summary()
```

Dla utworzonego powyej modelu, ksztat wyjciowy z pierwszej warstwy konwolucyjnej jest taki sam jak ksztat wejciowy (zastosowane zostao dopenienie zerami), a gboko wyjciowa jest rwna liczbie ustawionych filtrw - 2. 

W kolejnym etapie ustawmy parametry treningu i rozpocznijmy uczenie modelu na danych treningowych.

```python
# Ustawiamy parametry treningu

keras_model.compile(
    # wybr optymalizatora
    optimizer='adam',
    loss='categorical_crossentropy',
    # miara, ktr chcemy monitorowa
    metrics=['accuracy']
)

# Rozpoczynamy trening z zdefiniowanymi wczeniej hiperparametrami 
history = keras_model.fit(x_train, y_train, epochs=epochs, batch_size=batch_size)
```

Uzyskany wynik dokadnoci nie jest zadowalajcy, jednak stworzony model jest bardzo prosty i trenowany by przez 15 epok. Wywietlmy wykres z historii, aby zobaczy jak zmieniaa si dokadno modelu.

```python
figure = plt.figure()
figure.suptitle("Zmiana dokadnoci w trakcie treningu")
figure_acc = figure.add_subplot(1,1,1)
figure_acc.set_xlabel('epoka')
figure_acc.set_ylabel('miara - dokadno')
plt.plot(history.history['accuracy'])

plt.show()
```

Sprawdmy, jakie wartoci metryk ewaluacyjnych osiga wytrenowany model dla danych testowych.

```python
# Ewaluacja modelu na danych testowych
loss, accuracy = keras_model.evaluate(x_test, y_test)

print("Wynik ewaluacji - loss: ", loss)
print("Wynik ewaluacji - dokadno (accuracy): ", accuracy)
```
